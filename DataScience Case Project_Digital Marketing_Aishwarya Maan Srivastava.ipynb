Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Importing dataset

```python
df=pd.read_csv('online_advertising_performance_data.csv')
df
```

```python
# 'High' with 3
# 'Low' with 1
# 'Medium'  with 2

df['user_engagement'] = df['user_engagement'].replace({'High': 1, 'Low': 0, 'Medium': 0.5})
df
```

2.	How does the size of the ad (banner) impact the number of clicks generated?

```python
# Change the banner size from string to the numerical value which the string is actually expressing

def convert_banner_size(size_string):
  """
  This function takes a string representing a banner size and converts it to the numerical value it represents.
  """
  width, height = size_string.split(' x ')
  return int(width) * int(height)

df['banner_size'] = df['banner'].apply(convert_banner_size)
df.drop(['Unnamed: 12'],axis=1)
df.drop(['Unnamed: 13'],axis=1)
df
```

```python
# Plot of size of banner and clicks using matplotlib.pyplot

# Create a scatter plot with size of banner on the x-axis and clicks on the y-axis
plt.scatter(df['banner'], df['clicks'])

# Set the title and axis labels
plt.title('Banner size vs Clicks')
plt.xlabel('Banner size')
plt.ylabel('Clicks')
plt.xticks(rotation=90)
# Show the plot
plt.show()
```

4.	Is there a correlation between the cost of serving ads and the revenue generated from clicks?

12.	Are there any outliers in terms of cost, clicks, or revenue that warrant further investigation?

```python
# Calculate the z-scores for cost, clicks, and revenue
z_scores = df[['cost', 'clicks', 'revenue']].apply(lambda x: (x - x.mean()) / x.std())

# Identify outliers based on a threshold (e.g., z-score > 3)
outliers = z_scores[(z_scores['cost'] > 3) | (z_scores['clicks'] > 3) | (z_scores['revenue'] > 3)]

# Print the outliers
print("Outliers:")
print(outliers)

# Investigate the outliers further to understand the underlying reasons for their behavior
```

14.	Are there any specific campaigns or banner sizes that consistently outperform others in terms of ROI?

16.	Are there any noticeable differences in user engagement levels between weekdays and weekends?

```python
# Concatenate month and day with a valid separator
df['year'] = 2020
df['date'] = pd.to_datetime(df['month'].astype(str) + '/' + df['day'].astype(str) + '/' + df['year'].astype(str), format='%B/%d/%Y', dayfirst=True)
# Check the data types and values in the 'date' column
print(df['date'].dtypes)
print(df['date'].describe())
```

```python
# Create a new column to store the day of the week
df['day_of_week'] = pd.to_datetime(df['date']).dt.day_name()

# Calculate the average user engagement for each day of the week
avg_engagement_by_day = df.groupby('day_of_week')['user_engagement'].mean()

# Plot the average user engagement for each day of the week
avg_engagement_by_day.plot(kind='bar')
plt.title('Average User Engagement by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Average User Engagement')
plt.show()
```
