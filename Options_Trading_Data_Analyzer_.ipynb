```python
#convert date function
from datetime import datetime # Import the datetime module

def convert_date(row):
  return datetime.fromisoformat(row.datetime).strftime('%Y-%m-%d %H:%M:%S')
```

```python
#call conver date function
data['datetime'] = data.apply(convert_date, axis = 1)
```

```python
# get historical data function
def getdata(symbol, interval, date_from , date_to):
  inst_key = getinstrumentkey(symbol)
  url = f'https://api.upstox.com/v2/historical-candle/{inst_key}/{interval}/{date_to}/{date_from}'
  headers = {
      'Accept': 'application/json'
  }
  response = requests.get(url, headers=headers)
  # Check the response status
  if response.status_code == 200:
      # Do something with the response data (e.g., print it)
      print(response.json())
  else:
      # Print an error message if the request was not successful
      print(f"Error: {response.status_code} - {response.text}")

  data['datetime'] = data.apply(convert_date, axis = 1)
  data['datetime'] = pd.to_datetime(data['datetime'])
  return data
print(getdata('NIFTY24JUN22550PE', '1minute', '2024-05-24', '2024-05-31'))
```

```python
def get_option_chain_data(df: pd.DataFrame, side: str) -> pd.DataFrame:
    # Filter the data based on the 'side' argument
    if side == "PE":
        # Filter for Put options and find the highest bid price
        result = (df[['strike_price', 'put_bid_price']]
                    .groupby('strike_price')
                    .max()
                    .rename(columns={'put_bid_price': 'bid/ask'}))
    elif side == "CE":
        # Filter for Call options and find the highest ask price
        result = (df[['strike_price', 'call_ask_price']]
                    .groupby('strike_price')
                    .max()
                    .rename(columns={'call_ask_price': 'bid/ask'}))
    else:
        raise ValueError("Side must be either 'PE' or 'CE'")

    # Add the instrument name and side to the result
    result = result.reset_index()
    result['instrument_name'] = "NIFTY"  # You can customize this as needed
    result['side'] = side
    return result

```

```python
def calculate_margin_and_premium(data: pd.DataFrame, lot_size: int) -> pd.DataFrame:
    # Calculate premium earned
    data['premium_earned'] = data['bid/ask'] * lot_size

    # Simulate margin calculation (replace with API call if available)
    data['margin_required'] = 1000  # Placeholder value for margin requirement

    return data
```

```python
# Assuming df is the main DataFrame with all option data
side = "PE"  # Example: Put options
lot_size = 75  # Example lot size, replace with actual lot size

# Step 1: Get option chain data
options_data = get_option_chain_data(df, side)

# Step 2: Calculate margin and premium
final_data = calculate_margin_and_premium(options_data, lot_size)

# Display the final result
print(final_data)
```
