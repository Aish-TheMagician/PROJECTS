```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Accident_train.csv')
df
```

```python
for i in range(17):
    print("Number of null values in column",i,"are")
    nullcheck = df.iloc[:,i].isnull().sum()
    print(nullcheck)
```

```python
df=df.dropna()
df
```

```python
df.Hour_of_Collision.unique()
```

```python
df.Junction_Control.unique()
```

```python
df= df.reset_index(drop=True)
df
```

```python
dummies1=pd.get_dummies(df['Weekday_of_Collision'])
dummies1
```

```python
dummies2=pd.get_dummies(df['Policing_Area'])
dummies2
```

```python
df.drop(['Weekday_of_Collision', 'Policing_Area'],axis='columns', inplace=True)
df=pd.concat([df,dummies1,dummies2],axis='columns')
df
```

```python
Y=df['Collision_Severity']
Y.unique()
```

```python
X=df.drop(['Collision_Severity'],axis=1)
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=10)
```

```python
X_train
```

```python
X_test
```

```python
Y_train
```

```python
Y_test
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(X_train,Y_train)
Y_predicted=dt.predict(X_test)
print(Y_predicted)
print(np.array(Y_train[0:5]))
print(dt.score(X_test,Y_test))
```

```python
from sklearn import metrics
matrix=metrics.confusion_matrix(Y_predicted,Y_test)
sns.heatmap(matrix,cmap='Oranges',annot=True,fmt='0.1f')
plt.ylabel('Test values')
plt.xlabel('Predicted values')
plt.title('Confusion matrix for Decision Tree Clasifier')
plt.show()
```

```python
from sklearn.model_selection import cross_val_score
dt.fit(X,Y)
print(cross_val_score(dt,X,Y,cv=10))
print(np.average(cross_val_score(dt,X,Y,cv=10)))
```
