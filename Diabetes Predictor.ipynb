```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('diabetes.csv')
df
```

```python
Y = df['Outcome']
Y
```

```python
X = df.drop(['Outcome'],axis='columns')
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,train_size=0.8,random_state=7)
```

```python
Y_train
```

```python
Y_train
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,Y_train)
Y_predicted = lr.predict(X_test)

for i in range(len(Y_predicted)):
    if Y_predicted[i]<0.5:
        Y_predicted[i]=0
    else :
        Y_predicted[i]=1

print("The first five predcted values are:")
print(Y_predicted[0:5])
print("The first five test values are:")
print(np.array(Y_test[0:5]))
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_test, Y_test)
Y_predicted = knn.predict(X_test)
Y_predicted
```

```python
from sklearn import metrics
matrix = metrics.confusion_matrix(Y_predicted,Y_test)
sns.heatmap(matrix,annot=True,cmap="Blues",fmt='0.1f')
plt.xlabel("Predicted values")
plt.ylabel("Test values")
plt.title("Confusion matrix for KNeighborsClassifier")
plt.show()
```
